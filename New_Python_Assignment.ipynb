Additionally, the presence of invalid color categories highlights the need for data cleaning before drawing definitive conclusions.

```python
def remove_outliers_iqr(group):
    Q1 = group["sellingprice"].quantile(0.25)
    Q3 = group["sellingprice"].quantile(0.75)
    IQR = Q3 - Q1
    return group[(group["sellingprice"] >= Q1 - 1.5 * IQR) & (group["sellingprice"] <= Q3 + 1.5 * IQR)]

df_no_outliers = df.groupby("color").apply(remove_outliers_iqr)

plt.figure(figsize=(10,5))
sns.boxplot(x="color", y="sellingprice", data=df_no_outliers)
plt.xticks(rotation=90)
plt.xlabel("Color")
plt.title("Selling Price by Color (Outliers Removed)")
plt.tight_layout()

```

Outlier Removal:

Outliers in selling prices were removed using the Interquartile Range (IQR) method, applied separately within each color category. For each color, the 25th percentile (Q1) and 75th percentile (Q3) of selling prices were calculated, and values lying below Q1  1.5  IQR or above Q3 + 1.5  IQR were treated as outliers and excluded. This approach ensures that extreme prices are identified relative to each colors price distribution, allowing a fair comparison across colors without being distorted by rare luxury vehicles.

Insights from Selling Price by Color (Outliers Removed) graph:

1. Typical price range becomes clear: After removing outliers, most car prices fall within a limited, realistic range instead of being stretched by very expensive cars. The original data contained a few luxury cars with very high prices which distorted the data. These high price car data stretched the y-axis due to which the price range where most of the cars fall was not visible which is now clear. Same for median.

2. Neutral colors retain higher value: Colors such as white, black, gray, and silver show higher median selling prices, suggesting stronger buyer preference and better resale value.

3. Popular colors show more variability: Widely used colors have broader price distributions, reflecting availability across multiple vehicle segments.

4. Niche colors have stable but lower prices: Less common colors (e.g., green, gold, turquoise) exhibit narrower spreads and generally lower median prices.

5. Trends are robust: Even after removing outliers, the relative ordering of colors by median selling price remains consistent, confirming that insights are not driven by extreme values.
